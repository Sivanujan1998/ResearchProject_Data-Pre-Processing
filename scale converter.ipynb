```python
# define a function to map ages to a 5-point scale
def age_to_scale(age):
    if age < 20:
        return 1
    elif age < 23:
        return 2
    elif age < 26:
        return 3
    elif age < 30:
        return 4
    else:
        return 5
    
# 2scale to 5scale converter
def convert_2_to_5(value):
    # use a simple formula to map the value from the 2-scale to the 5-scale
    return int((value - 1) * (4/1) + 1)


# Define a function to convert a 3-point scale to a 5-point scale
def convert_3_to_5(old_value):
    new_value = (old_value - 1) * (4/2) + 1
    return new_value
```

```python
gsd['age'] = gsd['age'].apply(age_to_scale)
gsd['gender'] = gsd['gender'].apply(convert_2_to_5)
gsd['working_culture'] = gsd['working_culture'].apply(convert_3_to_5)
gsd['has_office'] = gsd['has_office'].apply(convert_2_to_5)
gsd['is_pegged'] = gsd['is_pegged'].apply(convert_2_to_5)
gsd['is_covid_gsd'] = gsd['is_covid_gsd'].apply(convert_2_to_5)
gsd['is_health_gsd'] = gsd['is_health_gsd'].apply(convert_2_to_5)
gsd['is_economic_crisis_gsd'] = gsd['is_economic_crisis_gsd'].apply(convert_2_to_5)
gsd['is_soft_skills_gsd'] = gsd['is_soft_skills_gsd'].apply(convert_2_to_5)
gsd['is_paying_structure_gsd'] = gsd['is_paying_structure_gsd'].apply(convert_2_to_5)
gsd['is_tools_gsd'] = gsd['is_tools_gsd'].apply(convert_2_to_5)
gsd['is_from_company_gsd'] = gsd['is_from_company_gsd'].apply(convert_2_to_5)
gsd['is_network_speed_gsd'] = gsd['is_network_speed_gsd'].apply(convert_2_to_5)
gsd['is_ergonomics_gsd'] = gsd['is_ergonomics_gsd'].apply(convert_2_to_5)
gsd['is_training_gsd'] = gsd['is_training_gsd'].apply(convert_2_to_5)
gsd['is_work_pressure_gsd'] = gsd['is_work_pressure_gsd'].apply(convert_2_to_5)
gsd['is_distraction_gsd'] = gsd['is_distraction_gsd'].apply(convert_2_to_5)
gsd['is_tech_skill_gsd'] = gsd['is_tech_skill_gsd'].apply(convert_2_to_5)
gsd['is_language_gsd'] = gsd['is_language_gsd'].apply(convert_2_to_5)
gsd['is_time_gsd'] = gsd['is_time_gsd'].apply(convert_2_to_5)
gsd['is_control_gsd'] = gsd['is_control_gsd'].apply(convert_2_to_5)
gsd['is_culture_gsd'] = gsd['is_culture_gsd'].apply(convert_2_to_5)
gsd['is_trust_gsd'] = gsd['is_trust_gsd'].apply(convert_2_to_5)
gsd['is_distance_gsd'] = gsd['is_distance_gsd'].apply(convert_2_to_5)
gsd['is_knowledge_gsd'] = gsd['is_knowledge_gsd'].apply(convert_2_to_5)
```
