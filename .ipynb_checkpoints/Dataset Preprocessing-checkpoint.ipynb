# Onehot encoding to crate dummy variables

```python
import sklearn.model_selection as ms
from sklearn import linear_model
from sklearn import preprocessing
import numpy as np
import numpy.random as nr
```

```python
def encode_variable(categorical_column):
    encoder = preprocessing.LabelEncoder()
    encoder.fit(categorical_column)
    encoded_categorical_column = encoder.transform(categorical_column)
    onehotencoder = preprocessing.OneHotEncoder()
    onehotencoder.fit(encoded_categorical_column.reshape(-1,1))
    return onehotencoder.transform(encoded_categorical_column.reshape(-1,1)).toarray()
```
